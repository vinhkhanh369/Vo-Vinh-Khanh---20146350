```python
import pandas as pd 
url = '/content/1.01. Simple linear regression.csv'
data = pd.read_csv(url)
data.head()

import seaborn as sns 
import matplotlib.pyplot as plt 
data.hist()
plt.show()

correlation = data.corr()
print (correlation)

sns.pairplot(data)

sns.heatmap(correlation, annot = True)
```

```python
data.plot(kind = 'scatter', x='SAT', y = 'GPA',figsize = (9,9),color='blue')

from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['SAT']),y = data['GPA'])
print('Gia tri Intercept la:',model.intercept_)
print('Gia tri Coeffient la:',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X = pd.DataFrame(data['SAT']),y = data['GPA']))
```

```python
data.plot(kind = 'scatter', x='SAT', y = 'GPA',figsize = (9,9),color='blue')

y_pred = model.predict(X = pd.DataFrame(data['SAT']))
plt.plot(data['SAT'],y_pred,color = 'red')
```
