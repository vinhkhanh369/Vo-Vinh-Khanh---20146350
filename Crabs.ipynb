```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
url = '/content/crabs.csv'
data = pd.read_csv(url)
data.hist()
plt.show
```

```python
correlation = data.corr() 
print(correlation)
```

```python
import seaborn as sns 
sns.pairplot(data)
```

```python
sns.heatmap(correlation,annot=True)
```

```python
data.plot(kind='scatter',x='depth',y='fontallip',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
# bac 1 
X=pd.DataFrame(data['depth'])
y=data['fontallip']
model.fit(X,y)

print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y))
```

```python
import numpy as np
data.plot(kind='scatter',x='depth',y='fontallip',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['depth'].min(),data['depth'].max(),0.1)
predictor = pd.DataFrame([range]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()

X=pd.DataFrame([data['depth'],data['depth']**2]).T
y=data['fontallip']
model.fit(X,y)

print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y))

import numpy as np
data.plot(kind='scatter',x='depth',y='fontallip',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['depth'].min(),data['depth'].max(),0.1)
predictor = pd.DataFrame([range,range**2]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()

X=pd.DataFrame([data['depth'],data['depth']**2,data['depth']**3]).T
y=data['fontallip']
model.fit(X,y)

print('Intercept cua mo hinh :',model.intercept_)
print('Coefficient cua mo hinh :',model.coef_)
print('Do chinh xac cua mo hinh:',model.score(X,y))

import numpy as np
data.plot(kind='scatter',x='depth',y='fontallip',figsize=(9,9),color = 'red',xlim=(0.7))
range =np.arange(data['depth'].min(),data['depth'].max(),1)
predictor = pd.DataFrame([range,range**2,range**3]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
